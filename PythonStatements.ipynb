If, Elif and Else Statements

```python
hungry = True

if hungry:
    print('FEED ME!')
else:
    print('Im not hungry')
```

```python
loc = 'Game'

if loc == 'Auto Shop':
    print("Cars are cool!")
elif loc == 'Bank':
    print("Money is cool!")
elif loc == 'Store':
    print("Welcome to the store!")
else:
    print("I do not know much.")
```

```python
name = 'Sara'

if name == 'Frankie':
    print("Hello Frankie!")
elif name == 'Sammy':
    print("Hello Sammy")
else:
    print("What is your name?")
```

List Comprehensions

```python
mystring = 'hello'
```

```python
mylist = []

for letter in mystring:
    mylist.append(letter)
```

```python
mylist
```

```python
#List comprehension is another way of flattening out a for loop
mylist = [letter for letter in mystring]
```

```python
mylist
```

```python
mylist = [x for x in 'word']
```

```python
mylist
```

```python
mylist = [num**2 for num in range(0,11)]
```

```python
mylist
```

```python
mylist = [x**2 for x in range(0,11) if x%2==0]
```

```python
mylist
```

```python
celcius = [0,10,20,34.5]

fahrenheit = [( (9/5)*temp + 32) for temp in celcius]
```

```python
fahrenheit
```

```python
fahrenheit = []

for temp in celcius:
    fahrenheit.append(( (9/5)*temp + 32))
```

```python
fahrenheit
```

```python
#if else within a list comprehension should only be used if it still allows your code to be readable
results = [x if x%2==0 else 'ODD' for x in range(0,11)]
```

```python
results
```
