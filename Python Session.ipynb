#### String Formatting
String formatting lets you inject items into a string rather than trying to chain items together using commas or string concatenation. 

There are various ways to perform string formatting.
- The oldest method involves placeholders using the modulo `%` character.
- Concatenation similar to what we do in SQL
- An improved technique uses the `.format()` string method.
- The newest method, introduced with Python 3.6, uses formatted string literals, called *f-strings*.

```python
player = 'Andy'
goals = 2

print('Yesterday, %s scored %s goals.' %())     # Usage of the Modulo operator %

print('Yesterday, ' + player + ' scored ' + goals + ' goals.')  # Concatenation similar to SQL

print('{}, {} {} {} {}'.format()) # Using the .format() method
      
print(f'Yesterday, {} scored {} goals.')         # f-string literals
```

`Let us now take it one step further and return the even numbers within a list`

```python
def check_even_list(num_list):
    
    even_numbers = []
    
    # Go through each number
    for number in num_list:
        # Once we get a "hit" on an even number, we append the even number
        if number % 2 == 0:
            even_numbers.append(number)
        # Don't do anything if its not even
        else:
            pass
    # Notice the indentation! This ensures we run through the entire for loop    
    return even_numbers
```

#### b. LESSER OF TWO EVENS: Write a function that returns the lesser of two given numbers *if* both numbers are even, but returns the greater if one or both numbers are odd
    lesser_of_two_evens(2,4) --> 2
    lesser_of_two_evens(2,5) --> 5

```python
# Define function using def keyword. Think about how you can find out if a number is even
def lesser_of_two_evens(a,b):

    if a % 2 == 0 and b % 2 == 0:
        return min(a,b)
    else:
        return max(a,b)
```

#### d. OLD MACDONALD: Write a function that capitalizes the first and fourth letters of a name
     
    old_macdonald('macdonald') --> MacDonald
    
Note: `'macdonald'.capitalize()` returns `'Macdonald'`

```python
# Define function using def keyword.
# Think about using index positions and the python built-in function "split()"
def old_macdonald(name):
    
    first_letter = name[0]
    inbetween = name[1:3]
    fourth_letter = name[3]
    rest = name[4:]
    
    return first_letter.upper() + inbetween + fourth_letter.upper() + rest
```

```python
# Check
old_macdonald('macdonald')
```

```python
# Can you write a more simplified version using first half and second half of MacDonald?
def old_macdonald(name):
    
    first_half = name[:3]
    second_half = name[3:]
    
    return first_half.capitalize() + second_half.capitalize()
```

```python
# Check
old_macdonald('macdonald')
```

#### e. Master Yoda: Given a sentence, return a sentence with the words reversed

    master_yoda('Are you Jedi powerful') --> 'powerful Jedi you are'
    master_yoda('Must I fight') --> 'fight I must'
    
Note: The .join() method may be useful here. The .join() method allows you to join together strings in a list with some connector string.

This means if you had a list of words you wanted to turn back into a sentence, you could just join them with a single space string. For example, the postcode strings below:

    >>> " ".join(['BD23','1AP'])
    >>> "BD23 1AP"

```python
# Define function using def keyword.
# Think about how you could use the split and join python methods
def master_yoda(text):
    
    wordlist = text.split()
    reverse_wordlist = wordlist[::-1]
    return ' '.join(reverse_wordlist)
```

#### h. Optimus Prime
Write a function that returns the number of prime numbers that exist up to and including a given number
    count_primes(100) --> 25
    
This is the first function that we will create that utilises a while loop

```python
# Define your function
def count_primes(num):
    
    # Logic for when the number is less than 2
    if num < 2:
        return 0
    
    # The prime numbers will always start at 2
    # Consider creating a placeholder list
    primes = [2]
    
    # Parameter to start at 3
    x = 3
    
    # Logic for counting the amount of prime numbers
    while x <= num:
        for y in primes:
            if x % y == 0:
                # This is just a simpler way of putting x = x + 2
                x += 2
                break
        else:
            primes.append(x)
            # This is just a simpler way of putting x = x + 2
            x += 2
    print(primes)
    return len(primes)
```

```python
count_primes(100)
```

```python
# Import the clear_output method from the IPython.display module
from IPython.display import clear_output

# Lets create the display board
def display_board(board):
    clear_output()
    
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
```

```python
# Lets create a function to mirror player input and decide who will be X or O
def player_input():
    marker = ''
    
    # If a marker has not been set then ask player 1 to select a marker
    while not (marker == 'X' or marker == 'O'):
        # Using .upper() will always uppercase the input by the user
        marker = input('Player 1: Pick X or O as your marker').upper()
        
    if marker == 'X':
        return ('X','O')
    else:
        return ('O','X')
```

```python
# Now let's create the function that checks for a winning combination
# Think of the different ways that you can win in Noughts and Crosses

# top row (7,8,9)
# middle row (4,5,6)
# bottom row (1,2,3)
# down left side (7,4,1)
# down middle (8,5,2)
# down right side (9,6,3)
# diagonal left (1,5,9)
# diagonal right (7,5,3)

def win_check(board,mark):
    
    return ((board[7] == mark and board[8] == mark and board[9] == mark) or # top row (7,8,9)
    (board[4] == mark and board[5] == mark and board[6] == mark) or # middle row (4,5,6)
    (board[1] == mark and board[2] == mark and board[3] == mark) or # bottom row (1,2,3)
    (board[7] == mark and board[4] == mark and board[1] == mark) or # down left side (7,4,1)
    (board[8] == mark and board[5] == mark and board[2] == mark) or # down middle (8,5,2)
    (board[9] == mark and board[6] == mark and board[3] == mark) or # down right side (9,6,3)
    (board[1] == mark and board[5] == mark and board[9] == mark) or # diagonal left (1,5,9)
    (board[7] == mark and board[5] == mark and board[3] == mark)) # diagonal right (7,5,3)
```

```python
# Lets see if X has won based on the test board we created earlier on
win_check(test_board,'X')
```
